```python
from pypdf import PdfReader

BASE_PATH="D:/Kaggle/MDC/train/PDF"

reader = PdfReader(f"{BASE_PATH}/10.1002_esp.5090.pdf")
text = ""
count=0
for page in reader.pages:
    text += page.extract_text()
    count+=1
print(text)
```

```python
cleaned_text = text.replace('\n', ' ')
cleaned_text = cleaned_text.replace('https://doi. ', 'https://doi.')
cleaned_text = cleaned_text.replace('/ ', '/')
cleaned_text = cleaned_text.replace('. ', '.')
cleaned_text = cleaned_text.replace(').', ' )')
cleaned_text = cleaned_text.replace(') ', ' )')
cleaned_text = cleaned_text.replace(',', ' ,')
cleaned_text = cleaned_text.replace('.https://doi.', '. https://doi.')
print(cleaned_text)
```

```python
# url_pattern = r'https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+'
# url_pattern = r'https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+(?:/[^\s]*)?'
url_pattern = r"https://doi\.org/10\.\d+/\S+"
# url_pattern = r'https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+(?:/[-\w.()]+)*'
extracted_url=re.findall(url_pattern, cleaned_text)
# cleaned_url=extracted_url.replace(
for url in extracted_url:
    print(url)
```

```python
i=-1
for url in extracted_url:
    i+=1
    if url =="https://doi.org/10.5066/p9353101":
        print(url,i)
```

```python

target_url = "https://doi.org/10.17882/49388"

# The long text string to search within
# long_text = t"""
# This is the first sentence. The second sentence is about oceanographic data,
# which can be found in the repository located at https://doi.org/10.17882/49388.
# This is the final sentence of the paragraph.
# """

# The name of the CSV file to append to
csv_file_path = 'extracted_sentences.csv'


# --- 2. Find the sentence containing the URL ---

# This regex finds a full sentence (text between periods) that contains your URL.
# re.escape() is used to handle special characters like '.' in the URL correctly.
pattern = r'([^.?!]*?' + re.escape(target_url) + r'[^.?!]*[.?!])'

match = re.search(pattern, text, re.IGNORECASE)


# --- 3. Append the sentence to the CSV file ---

if match:
    # Extract the matched sentence and remove leading/trailing whitespace
    sentence = match.group(1).strip()
    
    # Open the CSV file in append mode ('a')
    # newline='' is a best practice to prevent blank rows
    with open(csv_file_path, 'a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        
        # Write the sentence as a new row
        writer.writerow([sentence])
        
    print(f"Successfully found and appended sentence to '{csv_file_path}'")
    print(f"Appended sentence: '{sentence}'")
else:
    print(f"The URL '{target_url}' was not found in the text.")
```
